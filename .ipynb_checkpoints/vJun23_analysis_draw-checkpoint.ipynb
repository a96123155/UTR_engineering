```python
cao = pd.read_csv('Cao_HEK.csv')
cao = cao[['rpkm_riboseq', 'rpkm_rnaseq', 'te',
       'norm_riboseq', 'norm_rnaseq']]
cao = cao[(cao.rpkm_rnaseq > 5) & (cao.rpkm_riboseq > 0.1)]
cao = cao.dropna()
cao['log2TE'] = cao['te'].apply(np.log2)
cao
```

```python
data_1 = pd.read_excel('RVAC Preclinical RD_mRNA supply_m001-m100 sequence.xlsx')
data_1 = data_1[['mRNA ID', 'Category', 'Description',  "5' UTR"]]

data_1_pred = pd.read_csv('Design UTR_repredicted_m001-m100.csv')
data_1_pred = data_1_pred[['utr', 'Predicted log2TE',
       'Predicted_MRL', 'Method', 'Model', 'hek', 'hek_1', 'hek_CNN1',
       'hek_CNN2', 'pc3', 'muscle', 'egfp',]]
data_1 = pd.merge(data_1, data_1_pred, left_on = "5' UTR", right_on = 'utr')


# data_1[['Method', 'y_pred_raw']] = data_1['Description'].str.split(r'\s+', n=1, expand=True)
# data_1['y_pred'] = data_1['y_pred_raw'].str.extract(r'([\d\.]+)').astype(float)
# data_1.drop('y_pred_raw', axis=1, inplace=True)
data_1
```

```python
data_2 = pd.read_excel('20220516_100UTR prediction_m276-m375.xlsx')
data_2 = data_2[['utr', 'Predicted log2TE', 'Method', 'Predicted_MRL',
       'Model', 'hek1', 'hek2', 'hek_CNN1', 'hek_CNN2', 'pc3',
       'muscle', 'egfp']]
data_2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 设置更美观的样式
sns.set_style("white")  # 使用白色背景
# plt.rcParams['font.family'] = 'Arial'  # 设置字体
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

# 创建图形
fig, axes = plt.subplots(1, 2, figsize=(6, 3), dpi=300)
fig.patch.set_facecolor('white')  # 设置图形背景为白色

# 定义颜色
color_background = '#87CEEB'  # 天空蓝
color_design = '#FFA07A'      # 浅橙色

# 第一个子图: Optimus数据
ax1 = axes[0]
if len(optimus_background) > 0:
    sns.kdeplot(data=optimus_background, alpha=0.7, linewidth=2.5, 
                color=color_background, fill=True, ax=ax1,
                label=f'background (n={len(optimus_background):,})')
if len(gen_optimus) > 0:
    sns.kdeplot(data=gen_optimus, alpha=0.8, linewidth=2.5, 
                color=color_design, fill=True, ax=ax1,
                label=f'design (n=200)')

ax1.set_title('Optimus Distribution', fontsize=8, fontweight='bold', pad=10)
ax1.set_xlabel('RL (Ribosome Loading)', fontsize=6)
ax1.set_ylabel('Density', fontsize=5)
ax1.legend(fontsize=5, frameon=True, facecolor='white', edgecolor='none', framealpha=0.8)
ax1.grid(True, alpha=0.2, linestyle='-', linewidth=0.5)
ax1.set_facecolor('#FAFAFA')  # 轻微的灰色背景

# 美化刻度
ax1.tick_params(axis='both', which='major', labelsize=5)
ax1.spines['bottom'].set_linewidth(1)
ax1.spines['left'].set_linewidth(1)

# 第二个子图: HEK数据
ax2 = axes[1]
if len(hek_background) > 0:
    sns.kdeplot(data=hek_background, alpha=0.7, linewidth=2.5, 
                color=color_background, fill=True, ax=ax2,
                label=f'background (n={len(hek_background):,})')
if len(gen_hek) > 0:
    sns.kdeplot(data=gen_hek, alpha=0.8, linewidth=2.5, 
                color=color_design, fill=True, ax=ax2,
                label=f'design (n=200)')

ax2.set_title('HEK Distribution', fontsize=8, fontweight='bold', pad=10)
ax2.set_xlabel('log2TE (Translation Efficiency)', fontsize=6)
ax2.set_ylabel('Density', fontsize=5)
ax2.legend(fontsize=5, frameon=True, facecolor='white', edgecolor='none', framealpha=0.8)
ax2.grid(True, alpha=0.2, linestyle='-', linewidth=0.5)
ax2.set_facecolor('#FAFAFA')  # 轻微的灰色背景

# 美化刻度
ax2.tick_params(axis='both', which='major', labelsize=5)
ax2.spines['bottom'].set_linewidth(1)
ax2.spines['left'].set_linewidth(1)

# 调整子图间距
plt.tight_layout(pad=1.5)


# 保存高质量图片
plt.savefig('design_mrl_log2te_kde_comparison_plots.pdf', dpi=300, bbox_inches='tight', 
           facecolor='white', edgecolor='none')
```
